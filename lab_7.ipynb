```python
# Write your code here for the expected value
import numpy as np
import scipy.special as sps
import matplotlib.pyplot as plt
```

```python
shape = 5.0
scale = 1.0
samples = np.random.gamma(shape, scale, 100000)
count, bins, ignored = plt.hist(samples, 50, density=True)
y = bins**(shape-1)*(np.exp(-bins/scale) / (sps.gamma(shape)*scale**shape))
plt.plot(bins, y, linewidth=2, color='r')  
plt.show()
```

```python
def f(x):
  return np.sin(np.exp(x)) + np.log(x)
```

```python
expected_val = np.mean(f(samples))
expected_val
```
