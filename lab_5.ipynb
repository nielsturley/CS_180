## Exercise 4: America
Quantify the relationship between sugar and ratings. 

Make a plot to visualize this relationship. Superimpose a best fit line (with seaborn) to describe the relationship ([Seaborn RegPlot Documentation]("https://seaborn.pydata.org/generated/seaborn.regplot.html"))

```python
# Make the plot for the data visualization and line of best fit here:
sns.regplot(x=dta.sugars, y=dta.rating, fit_reg=True)
```


Calculate a statistic describing the relationship between sugar and ratings  (i.e., correlation). Write a statement in plain English interpreting this statistic.

```python
# Calculate the statistic using this cell:
np.corrcoef(dta.sugars, dta.rating)
```

The correlation coefficient for sugars vs. rating is -.76, indicating a strong negative relationship between the two variables--as sugar content increases, rating decreases.


## Exercise 5: America Part 2
Make a few plots comparing the relationships of carbo, sugars, calories, protein, and fat with rating.

```python
# Write your code to compare the various variables with rating below:
fig, axes = plt.subplots(2, 3, figsize=(20,10))
sns.regplot(ax=axes[0,0], x=dta.sugars, y=dta.rating, fit_reg=True)
sns.regplot(ax=axes[0,1], x=dta.carbo, y=dta.rating, fit_reg=True)
sns.regplot(ax=axes[0,2], x=dta.calories, y=dta.rating, fit_reg=True)
sns.regplot(ax=axes[1,0], x=dta.protein, y=dta.rating, fit_reg=True)
sns.regplot(ax=axes[1,1], x=dta.fat, y=dta.rating, fit_reg=True)
axes[1,2].remove()
```

Of the variables carbo, sugars, calories, protein, and fat which has the strongest relationship with rating? Justify your answer.

The one with the strongest relationship with rating is probably sugars. Since the error bar is pretty slim throughout, the points on the sugar graph can pretty accurately predict what the rating will be for the cereal. As sugar goes up, rating goes down.

## Exercise 7: Its Hot and It's Cold
The type column has two values: H='hot' and C='cold'. What is the average rating by type?

```python
# Write your code to get the average rating by type of cereal below:
dta_cold = dta.loc[dta['type'] == 'C']
cold_average_rating = dta_cold['rating'].mean()
print(f"Cold cereal average rating: {cold_average_rating}")

dta_hot = dta.loc[dta['type'] == 'H']
hot_average_rating = dta_hot['rating'].mean()
print(f"Hot cereal average rating: {hot_average_rating}")
```

## Exercise 8: Captain Crunch the Numbers
Provide one additional insight from this dataset that you found interesting. Create at least one figure and explain why the insight was interesting to you.

```python
# Create the extra plot below:
dta_mfrs = dta['mfr'].unique()
dta_mfrs
```

```python
ratings_list = []
for mfr in dta_mfrs:
  mfr_rating_df = dta.loc[dta['mfr'] == mfr]
  mfr_rating = mfr_rating_df['rating'].mean()
  ratings_list.append(mfr_rating)
  print(f"Manufacturer: {mfr}\nRating: {mfr_rating}\n")
dta_mfrs_rating = pd.DataFrame({
    'mfr': dta_mfrs,
    'overall_rating': ratings_list
})
```

```python
sns.barplot(data=dta_mfrs_rating, x="mfr", y="overall_rating")
```

Write why it was interesting here
