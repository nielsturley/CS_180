# Exercise 1: Means over axes

```python
import numpy as np
```

```python
def mean_for_loop(array):
    averages_array = []
    for i in array:
        current_average = 0
        for j in i:
            current_average += j
        current_average = current_average / len(i)
        averages_array.append(round(current_average, 8))
    return averages_array


def mean_vectors(array):
    array = np.array(array)
    array = array.mean(axis=1)
    array = array.round(8)
    return array
```

```python
X =   [[ 0.9573075 ,  0.83877059,  1.51767182],
       [ 0.83782936, -0.88288659,  0.7371596 ],
       [-0.01786695,  0.83195016,  0.66394427],
       [-2.43495362, -0.5999003 , -1.40044824],
       [-0.67749399, -0.93229581,  0.4736418 ],
       [ 0.47934903,  0.90244159, -0.24246888],
       [ 1.00280715, -0.68745287, -1.08115248],
       [ 0.98143632,  0.77163722,  0.07051186],
       [ 1.10721349,  0.37520477,  0.54710493],
       [ 0.20813717, -0.56000733, -0.38301232]]
print("Mean using for loops\n")
arr = mean_for_loop(X)
print(*arr, sep="\n")
arr = []
print()
print("Mean using vectors\n")
arr = mean_vectors(X)
print(*arr, sep="\n")
```

# Exercise 2: Element-wise operations

```python
def addition_for_loop(first_array, second_array):
    added_array = []
    for i, j in zip(first_array, second_array):
        added_array_row = []
        for k, l in zip(i, j):
            added_array_row.append(round(k + l, 8))
        added_array.append(added_array_row)
    return added_array


def addition_vectors(first_array, second_array):
    first_array = np.array(first_array)
    second_array = np.array(second_array)
    addition_array = first_array + second_array
    addition_array = addition_array.round(8)
    return addition_array
```

# Exercise 3: Dot Product

```python
def dot_product_for_loop(first_array, second_array):
    dot_product = 0
    for i, j in zip(first_array, second_array):
        product = i * j
        dot_product += product
    return dot_product


def dot_product_vectors(first_array, second_array):
    first_array = np.array(first_array)
    second_array = np.array(second_array)
    return np.dot(first_array, second_array)
```

```python
a = [4, 2, 6, 2, 5, 7, 5, 8, 8, 4]
b = [0, 6, 15, 3, 4, 6, 10, 8, 8, 18]

print("Dot product using for loop")
product = dot_product_for_loop(a,b)
print(product)
product = 0
print()
print("Dot product using vectors")
product = dot_product_vectors(a,b)
print(product)
```

# Exercise 4: Matrix Multiplication

```python
def multiply_for_loop(first_array, second_array):
    multiply_array = []
    temp_dot_product_array = []
    for k in first_array:
        multiply_array_row = []
        for i in range(len(second_array[0])):
            for j in second_array:
                temp_dot_product_array.append(j[i])
            prod = dot_product_for_loop(k, temp_dot_product_array)
            multiply_array_row.append(prod)
        multiply_array.append(multiply_array_row)
    return multiply_array


def multiply_vectors(first_array, second_array):
    first_array = np.array(first_array)
    second_array = np.array(second_array)
    return np.matmul(first_array, second_array)
```

```python
A = [[1,2],
     [2,1]]
B = [[3],
     [1]]

print("Matrix multiply using for loop")
array = multiply_for_loop(A,B)
print(*array, sep="\n")
array = []
print()
print("Matrix multiply using vectors")
array = multiply_vectors(A,B)
print(*array, sep="\n")
```
